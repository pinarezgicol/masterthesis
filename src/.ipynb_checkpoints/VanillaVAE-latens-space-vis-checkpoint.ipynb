```python
LATENT_DIMS = [64]
latent_dim= 64
MAX_EPOCHS = 200
BATCH_SIZE = 128
LEARNING_RATE = 0.001
EXPERIMENT_COUNT = 3
DEVICE = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
HIDDEN_DIM = 512
ORIGINAL_DIM = 784
epsilon_std = 1.0
```

### read training data

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.reshape(60000, 28, 28, 1).astype('float32') / 255.
X_test = X_test.reshape(10000, 28, 28, 1).astype('float32') / 255.
```

```python
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)
```

```python
# encoder architecture
x = Input(shape=(ORIGINAL_DIM,))
encoder_h = Dense(HIDDEN_DIM, activation='relu')(x)
z_mean = Dense(latent_dim)(encoder_h)

# encoder to generate latent variables from input
encoder = Model(x, z_mean)

encoder.load_weights(r"C:\Users\pinar\OneDrive\Masa端st端\masterthesis\src\Generative Models\mnist\mnist_vanillavae\mnist_vanillavae_z_mean_{latent_dim}.h5".format(latent_dim=latent_dim))

X_train_enhanced = encoder.predict(X_train, batch_size=32)
X_test_enhanced = encoder.predict(X_test, batch_size=32)
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=0, learning_rate=70, n_iter=2000, n_iter_without_progress=400, verbose=1)
EX2D = tsne.fit_transform(X_train_enhanced)
```

```python
import matplotlib.pyplot as plt
```
