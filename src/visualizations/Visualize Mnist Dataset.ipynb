```python
from keras.datasets import cifar10
import numpy as np
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
initial_idx = np.array([],dtype=int)
for i in range(10):
    idx = np.random.choice(np.where(y_train==i)[0], size=1, replace=False)
    initial_idx = np.concatenate((initial_idx, idx))
    
for i in range(10):
    idx = np.random.choice(np.where(y_train==i)[0], size=1, replace=False)
    initial_idx = np.concatenate((initial_idx, idx))
```

```python
fig = plt.figure( figsize=(1.5*10,2*2))

a = 1
for i in initial_idx:
    plottable_image = np.reshape(X_train[i], (32, 32, 3))
    ax = fig.add_subplot(2, 10, a)
    ax.imshow(plottable_image, cmap='gray')
    a = a + 1
    plt.axis('off')
fig.savefig("cifar10_examples.pdf")
```
